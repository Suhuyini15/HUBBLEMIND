## ENCODING

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

## BINARY ENCODING

```python
bin_col=['Gender','family_history_with_overweight','FAVC','SMOKE','SCC']
```

```python
label = LabelEncoder()
```

```python
for column in bin_col:
    df[column+'_encoded'] = label.fit_transform(df[column])
```

```python
df = df.drop(columns = bin_col)
```

```python
df
```

## CATEGORICAL ENCOODING

```python
cat_col= ['CAEC','CALC','MTRANS','NObeyesdad']
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoded_data = encoder.fit_transform(df[cat_col])
```

```python
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(cat_col))
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.head(20)
```

```python
df.drop(columns =['CAEC','CALC','MTRANS'],inplace=True)
```

```python
df
```

## OUTLIER HANDLING 

```python
df['Weight_transform'] = np.log(df['Weight'])
df['Height_transform'] = np.log(df['Height'])
```

```python
sns.boxplot(data=df,x='Weight_transform')
plt.grid()
plt.show()
```

```python
sns.boxplot(data=df,x='Height_transform')
plt.grid()
plt.show()
```

## NORMALIZING VALUES


```python
df['Weight'] = (df["Weight_transform"] - df["Weight_transform"].min()) / (df["Weight_transform"].max() - df["Weight_transform"].min())
df['Height'] = (df["Height_transform"] - df["Height_transform"].min()) / (df["Height_transform"].max() - df["Height_transform"].min())
df['Age'] = (df["Age"] - df["Age"].min()) / (df["Age"].max() - df["Age"].min())
```

```python
sns.boxplot(data=df,x='Height_transform')
plt.grid()
plt.show()
```

```python
sns.boxplot(data=df,x='Weight_transform')
plt.grid()
plt.show()
```

```python
df.drop(columns=['Height_transform','Weight_transform'],inplace=True)
```

```python
df.to_csv('./Obesity_Data.csv')
```
